```python
train = 'train.csv'
test = 'test.csv'
```

```python
import pandas as pd

df = pd.read_csv(train, index_col='PassengerId')
df_test = pd.read_csv(test, index_col='PassengerId')
```

```python
drop_columns = ['Ticket']
passthrough_columns=['Pclass', 'SibSp', 'Parch']
target_column = ['Survived']

text_columns_manquant = ['Cabin']  
text_columns_sans_manquant = ['Name']

numeric_columns_manquant =['Age', 'Fare']
cat_manquant = ['Embarked']
cat_sans_manquant = ['Sex']
```

```python
X = df.drop('Survived', axis=1)
y = df['Survived']
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import make_column_transformer 
from sklearn.pipeline import make_pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import FunctionTransformer


#pipeline pour le preprocesing des colonnes catégorielles avec des valeurs manquantes
cat_manquant_pipeline = make_pipeline(SimpleImputer(strategy='most_frequent'), OneHotEncoder(handle_unknown='ignore'))


#fonction transformer custom : preprocess de cabin
def extract_cabin_letter(df):
    return pd.DataFrame(df.str[0])

FunctionTransformer(extract_cabin_letter)

#pipeline pour le preprocesing de cabin

preprocessor_cabin = make_pipeline(
    FunctionTransformer(extract_cabin_letter),
    SimpleImputer(strategy='constant', fill_value='missing'),
    OneHotEncoder(handle_unknown='ignore')
)

#make_column_transformer: permet de faire un preprocessing spécifique pour chaque colonne
preprocessor = make_column_transformer(
    (OneHotEncoder(handle_unknown='ignore'), cat_sans_manquant),
    (cat_manquant_pipeline, cat_manquant),
    (SimpleImputer(strategy='median'), numeric_columns_manquant),
    (CountVectorizer(), 'Name'),
    (preprocessor_cabin, 'Cabin'),
    ('passthrough', passthrough_columns),
    ('drop', drop_columns),
)

```

```python
# pipeline final

from sklearn.pipeline import Pipeline

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import RidgeClassifier
from sklearn.neighbors import KNeighborsClassifier as KNN

from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

from sklearn.model_selection import GridSearchCV
```

```python
#KNN

#grid search pour trouver les meilleurs hyperparamètres

knn = {}

knn['pipeline'] = Pipeline([
    ('prepocessing', preprocessor),
    ('knn', KNN())
])

#hyperparamètres à tester
knn['hyperparameters'] = {}
knn['hyperparameters']['knn__n_neighbors'] = [1, 3, 5, 7, 9]
knn['hyperparameters']['knn__weights'] = ['uniform', 'distance']

#grid search
knn['grid_search'] = GridSearchCV(estimator=knn['pipeline'],
                                  param_grid=knn['hyperparameters'],
                                  cv=5,
                                  scoring='accuracy')    

knn['grid_search'].fit(X, y)

knn['grid_search'].best_params_  

knn['grid_search'].best_score_                   


# cross_val_score(knn_pipe, X, y, cv=5, scoring='accuracy').mean()
 
# knn_pipe.fit(X, y)

# pred_knn = knn_pipe.predict(X)

# # % de bonne prédiction
# accuracy_score(y, pred_knn)
```

```python
#RIDGE




#grid search pour trouver les meilleurs hyperparamètres

ridge = {}

ridge['pipeline'] = Pipeline([
    ('prepocessing', preprocessor),
    ('ridge', RidgeClassifier())
])

#hyperparamètres à tester
ridge['hyperparameters'] = {}
ridge['hyperparameters']['ridge__alpha'] = [5, 5.1, 5.2]


#grid search
ridge['grid_search'] = GridSearchCV(estimator=ridge['pipeline'],
                                  param_grid=ridge['hyperparameters'],
                                  cv=5,
                                  scoring='accuracy')    

ridge['grid_search'].fit(X, y)

ridge['grid_search'].best_params_  

ridge['grid_search'].best_score_   
```

```python
#RANDOM FOREST


#grid search pour trouver les meilleurs hyperparamètres
rf = {}

rf['pipeline'] = Pipeline([
    ('prepocessing', preprocessor),
    ('rf', RandomForestClassifier())
])



#hyperparamètres à tester
rf['hyperparameters'] = {}
rf['hyperparameters']['rf__n_estimators'] = [300, 400, 500]
rf['hyperparameters']['rf__max_depth'] = [40]
rf['hyperparameters']['rf__min_samples_split'] = [2, 3, 4]




#grid search
rf['grid_search'] = GridSearchCV(estimator=rf['pipeline'],
                                  param_grid=rf['hyperparameters'],
                                  cv=5,
                                  scoring='accuracy')    

rf['grid_search'].fit(X, y)

#score
rf['grid_search'].best_score_  
```

```python
#meilleur hyperparamètre du meilleur modèle
rf['grid_search'].best_params_  
```

```python
#Pipeline final avec le meilleur modèle et les meilleurs hyperparamètres
pipeline_final = make_pipeline(preprocessor, RandomForestClassifier(n_estimators=400, max_depth=40, min_samples_split=2))


#on le fit sur les données d'entrainement
pipeline_final.fit(X, y)

#on prédit sur les données de test
pipeline_final.predict(df_test)
```
